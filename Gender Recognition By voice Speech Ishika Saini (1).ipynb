```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Insights and adding label to male/female

```python
a=df.label.value_counts()
print(a)
print("TOTAL: ",df.label.count())
```

```python
df.corr() #correlation between different input variables
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['label'] = lb.fit_transform(df['label'])
#df['label'].value_counts()
print(df)
```

### Splitting dataset into taining set and testing set

```python
y = df.label.values
x = df.drop(['label'],axis='columns')
from sklearn.model_selection import train_test_split                  #importing class for splitting
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state =50) #20% test data, 80% train data
algo_names=[]
algo_scores=[]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.head()
```

```python
x_train.tail()
```

## Applying Different Models 

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier(random_state = 50)
dec_tree.fit(x_train,y_train)
print("Decision tree classification score: ",dec_tree.score(x_test,y_test))           #accuracy
algo_names.append("Descision Tree")
algo_scores.append(dec_tree.score(x_test,y_test))
p1=dec_tree.predict(x_test)                                                           #predicted values
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rand_forest = RandomForestClassifier(random_state = 50)
rand_forest.fit(x_train,y_train)
print("Random forest classification score: ",rand_forest.score(x_test,y_test))
algo_names.append("Random Forest")
algo_scores.append(rand_forest.score(x_test,y_test))
p2=rand_forest.predict(x_test)
```

### SVM 

```python
from sklearn.svm import SVC
svm=SVC(random_state = 50)
svm.fit(x_train,y_train)
print("SVM classification score is :{}".format(svm.score(x_test,y_test)))
algo_names.append("SVM")
algo_scores.append(svm.score(x_test,y_test))
p3=svm.predict(x_test)
```

### KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
print("KNN classification score is: ",knn.score(x_test,y_test))
algo_names.append("KNN")
algo_scores.append(knn.score(x_test,y_test))
p4=knn.predict(x_test)
```

### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='liblinear')
lr.fit(x_train,y_train)
print("Logistic Regression classification score is: ",lr.score(x_test,y_test))
algo_names.append("Logistic Regression")
algo_scores.append(lr.score(x_test,y_test))
p5=lr.predict(x_test)
```

### Classification Report and Confusion Matrix

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
report_dec_tree = classification_report(y_test,p1)                          #decision tree
print(report_dec_tree)
```
